```python
def mycollect(e1, e2):
    """
    Collects e2 coefficients in e1, even when e2 is not a factor of
    some of the terms of e1
    """
    g = maxima_calculus.gensym()._sage_()
    while g in e1.variables(): g = maxima_calculus.gensym()._sage_()
    return sum([v[0]*g^v[1]
                for v in e1.subs(e2==g).coefficients(g)]).subs(g==e2)
```

```python
def ele_dict(deg):
        es=[e[i].expand(deg)*(-1)**i for i in range(1,deg+1)]
        coe=list(reversed([eval("a"+str(i)) for i in range(0,deg)]))
        return dict(zip(coe,es))
```

```python
def poly(n):
    return sum([x0^(j)*eval("a"+str(j)) for j in [0..n-1]])+x0^n
```

```python
def ordi(exp,vari):
    terms=exp.operands()
    lonely_terms=[i for i in terms if not(i.has(vari))]
    fixed=[i for i in terms if i not in lonely_terms]+[sum(lonely_terms)]
    return fixed
```

```python
def red(op,deg):
    po=poly(deg).operands()
    po[0:2]=list(reversed(po[0:2]))
    if deg>=4:
        for i in range(0,deg-3):
            leading=op[i].subs({x0:1})
            new=[leading*j*(x0^(deg-4-i)) for j in po]
            if -(deg-4-i)<0:
                op=[(x-y).full_simplify().factor() for x,y in zip(op[0:-(deg-4-i)],new)]+op[len(op)-(deg-4-i):len(op)]
            else:
                op=[(x-y).full_simplify().factor() for x,y in zip(op,new)]
            op=op[1:]
        return op
    else:
        return op
        
```

```python
def red2(op,deg):
    po=poly(deg)
    if deg>=4:
        for i in range(0,deg-3):
            leading=op.operands()[0].subs({x0:1})
            op=op-leading*(x0^(deg-4-i))*po
            op=mycollect(op.full_simplify(),x0)
        return op
    else:
        return op
```

```python
def asum(n):
    return sum([(j+1)*(j)/2*x0^(j-1)*eval("a"+str(j+1)) for j in [1..n-2]])+n*(n-1)/2*x0^(n-2)
def bsum(n):
    return sum([j*(n-j)*x0^(j-1)*eval("a"+str(j)) for j in [1..n-1]])
def csum(n):
    return sum([(n-j+1)*(n-j)/2*x0^(j-1)*eval("a"+str(j-1)) for j in [1..n-1]])
def csum_2(n):
    return sum([(j-1)*(j)/2*x0^(n-j)*eval("a"+str(n-j)) for j in [2..n]])
def csum_3(n):
    return (x0^n)*sum([(j-1)*(j)/2*x0^(-j)*((-1)^j)*e[j].expand(n) for j in [2..n]])
def inti(n):
    return sum([x0^(n-j)*eval("a"+str(n-j)) for j in [1..n]])
def csum_4(n):
    return sum([sum([x0^(n-j)*eval("a"+str(n-j)) for j in [1..i]])*-i for i in [1..n-1]])+((n)*(n-1)/2)*inti(n)
def qterm_2(n):
    return ((n)*(n-1)/2)*sum([((-1)^j)*q_binomial(n,j)*q^(j*(j-1)/2) for j in [1..n]])
def qterm_1(n):
    return sum([sum([((-1)^j)*q_binomial(n,j)*q^(j*(j-1)/2) for j in [1..i]])*-i for i in [1..n-1]])
def q_csum_2(n):
    return sum([(j-1)*(j)/2*((-1)^j)*q_binomial(n,j)*q^(j*(j-1)/2) for j in [2..n]])
```
