```python
import pandas as pd
import numpy as np
```

```python
all_table=pd.read_csv("output_all_galois_deg4_range25.csv")
```

```python
for ht in range(0,4):
    table_zeros=all_table[all_table["zero_terms"]==ht]
    print(table_zeros["galois_group"].value_counts())
```

```python
first_zero=all_table[all_table["term_0"]==0]
```

```python
first_zero["galois_group"].value_counts()
```

```python
nonfirst_zero=all_table[~(all_table["term_0"]==0)]
```

```python
nonfirst_zero["galois_group"].value_counts()
```

```python
cyclic_non_zero=nonfirst_zero[nonfirst_zero["galois_group"]=="C(4)"].copy().sort_values(["term_0","term_1","term_2", "term_3"])
klein_non_zero=nonfirst_zero[nonfirst_zero["galois_group"]=="E(4)"].copy().sort_values(["term_0","term_1","term_2", "term_3"])
dihedral_non_zero=nonfirst_zero[nonfirst_zero["galois_group"]=="D(4)"].copy().sort_values(["term_0","term_1","term_2", "term_3"])
```

```python
cyclic_non_zero.to_csv("cyclic_non_zero.csv",index=False)
klein_non_zero.to_csv("klein_non_zero.csv",index=False)
dihedral_non_zero.to_csv("dihedral_non_zero.csv",index=False)
```

```python
knon=klein_non_zero[klein_non_zero["zero_terms"]==0]
dnon=dihedral_non_zero[dihedral_non_zero["zero_terms"]==0]
cnon=cyclic_non_zero[cyclic_non_zero["zero_terms"]==0]
```
